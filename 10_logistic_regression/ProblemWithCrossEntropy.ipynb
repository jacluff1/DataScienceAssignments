```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.array([1,2,1]).reshape([1,3])
y = np.array([1])
```

```python
def sigmoid(h):
    return 1/(1 + np.exp(-h))

def cross_entropy(y, p_hat):
    return -np.sum(y*np.log(p_hat) + (1 - y)*np.log(1 - p_hat))

def accuracy(y, p_hat):
    return np.mean(y == np.round(p_hat))

# w = np.random.randn(4)
eta = 1e-3
epochs = int(1e3)
J = [0]*epochs

for epoch in range(epochs):
    w = np.array([1,epoch,epoch]).reshape([3,1])
    p_hat = sigmoid(X.dot(w))
    J[epoch] = cross_entropy(y, p_hat)
```

```python
plt.figure(figsize=(10,10))
plt.plot(np.arange(epochs), J)
plt.xlabel("w", fontsize=20)
plt.ylabel("J", fontsize=20)
plt.title("The lies", fontsize=20)
```
