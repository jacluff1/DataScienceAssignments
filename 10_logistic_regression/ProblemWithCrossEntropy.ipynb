```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.array([1,2,1]).reshape([1,3])
y = np.array([1])
```

```python
def sigmoid(h):
    return 1/(1 + np.exp(-h))

def cross_entropy(y, p_hat):
    return -np.sum(y*np.log(p_hat) + (1 - y)*np.log(1 - p_hat))

def accuracy(y, p_hat):
    return np.mean(y == np.round(p_hat))

# w = np.random.randn(4)
eta = 1e-3
epochs = int(1e3)
J = [0]*epochs

for epoch in range(epochs):
    w = np.array([1,epoch,epoch]).reshape([3,1])
    p_hat = sigmoid(X.dot(w))
    J[epoch] = cross_entropy(y, p_hat)
```

```python
plt.figure(figsize=(10,10))
plt.plot(np.arange(epochs), J)
plt.xlabel("w", fontsize=20)
plt.ylabel("J", fontsize=20)
plt.title("The Subjective Objective Function", fontsize=20)
```

First, lets look at the functions:

$h(x) = w^T x \longrightarrow \hat{p}(h) = \frac{1}{1 + e^{-h}} \longrightarrow J = -\sum_{i=1}^N y_i \ln\hat{p} + (1-y_i)\ln(1-\hat{p})$

Now, lets look at the limits of each function:
$$\lim_{w\to\infty} h(x) = \infty$$
$$\lim_{h\to\infty} \hat{p}(h) = 1$$
$$\lim_{p\to1} \ln\hat{p} = 0,\ \lim_{p\to1} \ln(1-\hat{p}) = -\infty$$

Overall, this implies that increasing $w$ will have the effect of decreasing $J$ This means that the basic model will tend to favor higher weight values, which means that the cross entropy objective function would be better named the subjective objective function.
